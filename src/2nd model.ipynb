```python
Datadirectory = "data/train/"
```

```python
Classes = ["angry","disgust","fear","happy","neutral","sad","surprise"] 
# 0 = angry, 1 = disgust, 2 = fear, 3 = happy, 4 = neutral, 5 = sad, 6 = surprise
```

```python
for category in Classes:
    path = os.path.join(Datadirectory, category)
    for img in os.listdir(path):
        img_array = cv2.imread(os.path.join(path,img))
        plt.imshow(cv2.cvtColor(img_array, cv2.COLOR_BGR2RGB))
        plt.show()
        break
    break
```

```python
img_size = 224
new_array = cv2.resize(img_array, (img_size,img_size))
plt.imshow(cv2.cvtColor(new_array, cv2.COLOR_BGR2RGB))
plt.show()
```

```python
new_array.shape
```

```python
training_Data = [] ## data

def create_training_Data():
    for category in Classes:
        path = os. path.join(Datadirectory, category)
        class_num= Classes. index (category) ## 0 1, ## Label
        for img in os.listdir(path):
            try:
                img_array = cv2.imread (os.path.join(path, img))
                new_array= cv2.resize(img_array, (img_size, img_size))
                training_Data.append([new_array,class_num])
            except Exception as e:
                pass
```

```python
create_training_Data()
```

```python
print(len(training_Data))
```

```python
import random
random.shuffle(training_Data)
```

```python
X = [] ## data /feature
Y = [] ## Label
for features, label in training_Data:
    X.append(features)
    Y.append(label)
    
X = np.array(X).reshape (-1, img_size, img_size, 3) ## converting it to 4 dimension
```

```python
X.shape
```

```python
# normalize
# X = X/255.0;
```

```python
type (Y)
```

```python
Y=np.array(Y)
```

```python
Y.shape
```

```python
from tensorflow import keras
from tensorflow.keras import layers
```

```python
model = tf.keras.applications.MobileNetV2() #PreTrained Model
```

```python
model.summary()
```

```python
base_input = model.layers[0].input
```

```python
base_output = model.layers[-2].output
```

```python
base_output
```

```python
final_output = layers.Dense(128)(base_output) ## adding new Layer, after the output of global pooling Layer
final_output = layers. Activation('relu')(final_output) ## activation function
final_output = layers. Dense (64)(final_output)
final_output = layers. Activation('relu')(final_output)
final_output = layers.Dense(7, activation='softmax')(final_output) ## my classes are 07
```

```python
final_output
```

```python
new_model = keras.Model(inputs = base_input, outputs= final_output)
```

```python
new_model.summary()
```

```python
new_model.compile(loss="sparse_categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
```

```python
new_model.fit(X,Y, epochs = 1)
```

```python
new_model.save('FERModelPrototype.h5')
```
